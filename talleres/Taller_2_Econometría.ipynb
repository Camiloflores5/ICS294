### 1.3 Cree una variable de la multuplicación entre la variable dummy antes creada y la variable indus. ¿Qué indica esta interacción?

```r

# Crear la variable de interacción
ceosal1$sales_mayor_promedio_indus <- ceosal1$sales_mayor_promedio * ceosal1$indus

# Mostrar las primeras filas del dataset con la nueva variable de interacción
cat("Primeras filas del dataset 'ceosal1' con la nueva variable 'sales_mayor_promedio_indus':\n")
print(head(ceosal1))

# Verificar la estructura de la nueva variable
cat("\nResumen estadístico de 'sales_mayor_promedio_indus':\n")
summary(ceosal1$sales_mayor_promedio_indus)
```

**Interpretación:**

Esta variable de interacción permite capturar el **efecto diferencial** o **efecto conjunto** de tener ventas por encima del promedio *dentro de la industria específica que representa `indus`*. En un modelo de regresión, por ejemplo:

*   Si `indus` es una dummy para una industria específica (ej. `indus = 1` para manufactura y `0` para otros), esta variable indicaría si una empresa está en la industria de manufactura *y* tiene ventas por encima del promedio.
*   Permite que el efecto de `sales_mayor_promedio` sobre la variable dependiente (por ejemplo, `salary`) sea diferente para las empresas en la industria `indus` en comparación con las empresas que no están en esa industria. O viceversa, que el efecto de pertenecer a la industria `indus` sea diferente para las empresas con ventas por encima del promedio.


### 1.4 Cree una variable de la multiplicación entre la variable continua roe y la variable dummy antes creada. ¿Qué indica esta interacción?


```r
# Crear la variable de interacción
ceosal1$sales_mayor_promedio_roe <- ceosal1$sales_mayor_promedio * ceosal1$roe

# Mostrar las primeras filas del dataset con la nueva variable de interacción
cat("Primeras filas del dataset 'ceosal1' con la nueva variable 'sales_mayor_promedio_roe':\n")
print(head(ceosal1))

# Verificar la estructura de la nueva variable
cat("\nResumen estadístico de 'sales_mayor_promedio_roe':\n")
summary(ceosal1$sales_mayor_promedio_roe)
```

### ¿Qué indica esta interacción?

La variable `sales_mayor_promedio_roe` es una **variable de interacción** entre una variable dummy (`sales_mayor_promedio`) y una variable continua (`roe`). Su propósito es capturar un **efecto diferencial** en la pendiente de `roe` o en el intercepto, dependiendo de cómo se utilice en un modelo de regresión.

Analicemos los valores que puede tomar `sales_mayor_promedio_roe`:

*   Si `sales_mayor_promedio` es `0` (las ventas no superan el promedio), entonces `sales_mayor_promedio_roe` será `0 * roe = 0`.
*   Si `sales_mayor_promedio` es `1` (las ventas superan el promedio), entonces `sales_mayor_promedio_roe` será `1 * roe = roe`.

**Interpretación:**

Esta interacción indica que el **efecto de `roe` sobre la variable dependiente (por ejemplo, `salary`) puede ser diferente para las empresas cuyas ventas superan el promedio en comparación con las empresas cuyas ventas no lo hacen.**

En un modelo de regresión lineal simple, si tuvieras una ecuación como:

`Y = β0 + β1*sales_mayor_promedio + β2*roe + β3*sales_mayor_promedio_roe + ε`

*   Para empresas con `sales_mayor_promedio = 0`:
    `Y = β0 + β2*roe + ε` (El efecto de `roe` es `β2`)

*   Para empresas con `sales_mayor_promedio = 1`:
    `Y = β0 + β1 + β2*roe + β3*roe + ε`
    `Y = (β0 + β1) + (β2 + β3)*roe + ε` (El efecto de `roe` es `β2 + β3`)

Por lo tanto, `β3` representa el **cambio en la pendiente** de `roe` cuando las ventas de la empresa superan el promedio. En otras palabras, la variable de interacción permite que el impacto del retorno sobre el capital (`roe`) en la variable dependiente sea diferente para los dos grupos definidos por `sales_mayor_promedio`.

*   **`ros`**: El `p-value` para `ros` es `0.0177`, que es menor que 0.05. Esto indica que `ros` es estadísticamente **significativo** al 5% para explicar el salario del CEO.

### 2.3 Realiza un test de hipotesis para evaluar si las dos variables que se agregaron son significativas para el modelo simultaneamente. Use significancia del 5%.


```r
F_critico <- qf(0.95 , 2 , 203)
R2_r <- 0.03191
R2_ur <- 0.04059
F_est <- (R2_ur - R2_r)/(1-R2_ur)*203/2
if(F_est > F_critico) {"Se rechaza H0"} else {"No se rechaza H0"}
```

### **Interpretación**

Esto implica que no tenemos suficiente evidencia estadística para concluir que las variables sales_mayor_promedio y sales_mayor_promedio_roe son conjuntamente significativas para explicar el salario del CEO.

### 2.4 Volviendo a la regresión del salario en función de sales, roe y ros. Considere dos grupos, Grupo 1: Aquellos clientes que sales es mayor al promedio, y Grupo 2: Aquellos clientes que sales es menor al promedio. Realice un test de hipotesis para grupos al 5% de significancia. Use el metodo que más le acomode.


```r
ceosal1$sales_D <- ceosal1$sales*ceosal1$sales_mayor_promedio
ceosal1$roe_D <- ceosal1$roe*ceosal1$sales_mayor_promedio
ceosal1$ros_D <- ceosal1$ros*ceosal1$sales_mayor_promedio

modelo <- lm(salary ~ sales+roe+ros+sales_mayor_promedio+sales_D+roe_D+ros_D, data=ceosal1)
summary(modelo)

F_critico <- qf(0.95, 4, 201)
R2_ur <- 0.06028
R2_r <- 0.03191
F_est <- (R2_ur-R2_r)/(1-R2_ur)*201/4
if (F_est > F_critico) {"Se rechaza H0"} else {"No se rechaza H0"}
```

### **Interpretación**
No se rechaza la Hipótesis Nula (H0) en el test de grupos, esto significa que, a un nivel de significancia del 5%, no encontramos evidencia estadística suficiente para concluir que existen diferencias significativas entre los dos grupos (empresas con ventas mayores al promedio y empresas con ventas no mayores al promedio) en cómo los regresores (sales, roe, ros) afectan el salario del CEO.

### 2.5  Usando la regresión del salario en función de sales, roe y ros, se distinguen los grupos mediante las dummies; indus, finance, consprod y utility, que corresponden al tipo de empresa. Realice mediante test de Chow, con 5% de significancia, si los modelos para cada grupo son iguales.

```r
ceosal1_indus <- subset(ceosal1, indus==1)
ceosal1_finance <- subset(ceosal1, finance ==1)
ceosal1_consprod <- subset(ceosal1, consprod==1)
ceosal1_utility <- subset(ceosal1, utility==1)

modelo <- lm(salary ~ sales+roe+ros, data=ceosal1)
modelo_indus <- lm(salary ~ sales+roe+ros, data=ceosal1_indus)
modelo_finance <- lm(salary ~ sales+roe+ros, data=ceosal1_finance)
modelo_consprod <- lm(salary ~ sales+roe+ros, data=ceosal1_consprod)
modelo_utility <- lm(salary ~ sales+roe+ros, data=ceosal1_utility)

SSR_r <- sum(residuals(modelo)^2)
SSR_indus <- sum(residuals(modelo_indus)^2)
SSR_finance <- sum(residuals(modelo_finance)^2)
SSR_consprod <- sum(residuals(modelo_consprod)^2)
SSR_utility <- sum(residuals(modelo_utility)^2)
SSR_ur <- SSR_indus+SSR_finance+SSR_consprod+SSR_utility

G <- 4
k <- 3
n <- nrow(ceosal1)
gl1 <- (G-1)*(k+1)
gl2 <- n-G*(k+1)

F_critico <- qf(0.95, gl1, gl2)
F_est <- (SSR_r-SSR_ur)/SSR_ur*gl2/gl1
if(F_est > F_critico) {"Se rechaza H0"} else {"No se rechaza H0"}
```

### **Interpretación**
No se rechaza la Hipótesis Nula (H0) en el test de Chow. Esto significa que, con un nivel de significancia del 5%, no hay evidencia estadística suficiente para concluir que los modelos de regresión del salario (en función de sales, roe y ros) son diferentes entre los grupos definidos por las variables dummy indus, finance, consprod y utility. En otras palabras, podemos asumir que la relación entre el salario del CEO y las variables sales, roe y ros es la misma para todas las categorías de empresas (industria, finanzas, productos de consumo y servicios públicos).
